# Univariate Analysis

```python
# To plot a histogram to dipict the status of people according to age.

sns.FacetGrid(hb, hue="status", size=5) \
   .map(sns.distplot, "age", kde=False, bins=5) \
   .add_legend();

```

```python
#----------Observations--------------

#-The Histogram has been plotted for Stauts that is 1- people living more than 5 years after surgery and 2- Pleople dying before 5 years
#according to age.
#-This means for every age range what is the count of people that fall under status 1 and status 2 respectively.
#-eg: for status 1 i.e. the graph in blue colour: for age range of 30-40 their are approx. 38 people that lie in Status 1 category.
#- Same way we can estimate for status 2 for which the graph is given in red in colour as well: For age group of 35-45 their are 
#  approx. 12 people that fall in status 2 category.
  
```

```python
hb_status1 = hb[hb["status"] == 1].copy();
hb_status2 = hb[hb["status"] == 2].copy();
```

```python
# To Plot the PDF and CDF for age by status classes '1' and '2'.


plt.title('pdf and cdf of age by status for both class 1 and 2')
counts, bin_edges = np.histogram(hb_status1['age'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf,'y',label='pdf_age1')
plt.plot(bin_edges[1:], cdf,'g',label='cdf_age1')
plt.legend()
counts, bin_edges = np.histogram(hb_status2['age'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf,'r',label='pdf_age2')
plt.plot(bin_edges[1:], cdf,'b',label='pdf_age2')
plt.legend()
plt.show();
```

```python
# To Plot the PDF and CDF for year by status classes '1' and '2'.

plt.title('pdf and cdf of year by status for both class 1 and 2')
counts, bin_edges = np.histogram(hb_status1['year'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf,'y',label='pdf_year1')
plt.plot(bin_edges[1:],cdf,'g',label='cdf_year1')
plt.legend()

counts, bin_edges = np.histogram(hb_status2['year'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf,'r',label='pdf_year2')
plt.plot(bin_edges[1:], cdf,'b',label='cdf_year2')
plt.legend()
plt.show();
```

```python
# To Plot the PDF and CDF for nodes by status classes '1' and '2'.

plt.title('pdf and cdf of nodes by status for both class 1 and 2')
counts, bin_edges = np.histogram(hb_status1['nodes'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf,'y',label='pdf_nodes1')
plt.plot(bin_edges[1:], cdf,'g',label='cdf_nodes1')
plt.legend()
counts, bin_edges = np.histogram(hb_status2['nodes'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf,'r',label='pdf_nodes2')
plt.plot(bin_edges[1:], cdf,'b',label='cdf_nodes2')
plt.legend()
plt.show();
```

```python
#---------------Observation from the PDF and CDF curves----------

# Approx. 80% of the people lived more than 5 years who had less than 10 Auxilliary nodes.
# Most number of operations were done between years 1965-1967 for people who lived less than 5 years.
# Most number of nodes were detected in the range of 20-30 for people who lived less than 5 years.
# Maximum number of people were operated in the year 1964 who live for more than 5 years.
# Maximum number of people who survived for more than 5 years are in the age group of 50-60.
```

# Bivariate Analysis

```python
# Scatter plot analysis.

sns.set_style("whitegrid");
sns.FacetGrid(hb, hue="status", size=5) \
   .map(plt.scatter, "age", "nodes") \
   .add_legend();
plt.show();
```

```python
#-------Observations---------

# From the scatter plot plotted with x-axis as age and y-axis as nodes is one of the logical attribute combinations for which the
# scatter plot has been made.
# But not a lot of details can be obtained from this plot as there are many overlapings.
# One thing that can be considered is that the person with maximum nodes i.e approx. 52 that belonged to status "1" has an age approx. 43.
# And the person with maximum nodes i.e approx. 47 that belonged to status "2" has an age approx. 45.
# For Scatter plots with other combinations please refer the pair plots below.
```

```python
# Pair Plot Analysis

sns.set_style("whitegrid");
sns.pairplot(hb, hue="status", size=3);
```

```python
#-------Observations---------
# Features age and year cannot be considerd for classification as they produce many overlaps and dont give any useful information.
# Plots shown in pair plots give massive overlaps so, they are not that user friendly for classification of patients according 
# to status for difernt attributes.
```

```python
#---------Conclusion---------

# Univarite analysis is more useful than bivariate analysis for classification.
# Univariate analysis tend to give more precise information as seen in box plots, violin plots, PDF and CDF which can help in 
# classification where as bivariate analysis gives overlaps as seen in scatter and pair plots as are not useful for classification.
```
